## 3. Preparar columnas del Dataset según la información que representan

```python
df["Fecha_cierre"] = pd.to_datetime(
    df["Fecha_cierre"],
    format="%d/%m/%y %H:%M:%S"
)
```

```python
df["Fecha_apertura"] = pd.to_datetime(
    df["Fecha_apertura"],
    format="%d/%m/%y %H:%M:%S"
)
```

```python
df["Aplicacion"] = df["Aplicacion"].replace({
    "FIRST DATA": "FIRST VISION",
    "SWIFT IZQUIERDO": "SWIFT",
    "SWIFT DERECHO": "SWIFT",
    "SERIVA RENTA FIJA": "SERIVA"
})

df["Aplicacion"] = df["Aplicacion"].fillna("OTRA APP")
```

```python
def cambiar_sitio(valor):
    if valor == "A":
        return "EDIFICIO PRINCIPAL"
    elif valor == "B":
        return "OFICINA BANCO"
    elif valor == "C":
        return "OFICINA BANCO"
    elif valor == "D":
        return "AREA ESPECIAL"
    else:
        return "EDIFICIO PRINCIPAL"

df["Sitio"] = df["Sitio"].apply(cambiar_sitio)
```

```python
diccionario_solicitudes = {
    "CREACIÓN": "CREACION",
    "CREACION": "CREACION",
    "CREAR": "CREACION",
    "ASIGNACION": "CREACION",
    "ASIGNACIÓN": "CREACION",
    "MODIFICACION": "MODIFICACION",
    "MODIFICACIÓN": "MODIFICACION",
    "CAMBIAR": "MODIFICACION",
    "ELIMINACIÓN": "ELIMINACION",
    "ELIMINAR": "ELIMINACION",
    "DESBLOQUEO": "DESBLOQUEO",
    "DESBLOQUEAR": "DESBLOQUEO",
    "ACTIVACION": "DESBLOQUEO",
    "ACTIVAR": "DESBLOQUEO",
    "RESETEO": "DESBLOQUEO",
    "CONTRASEÑA": "DESBLOQUEO",
    "PASSWORD": "DESBLOQUEO",
    "CLAVE": "DESBLOQUEO",
    "REACTIVACION": "DESBLOQUEO",
    "RECUPERACION": "DESBLOQUEO",
    "RECUPERACIÓN": "DESBLOQUEO",
    "RESET": "DESBLOQUEO"
}

for palabra, tipo in diccionario_solicitudes.items():
    mask = df["Titulo_caso"].str.contains(palabra, case=False, na=False)
    df.loc[mask, "Tipo_solicitud_1"] = tipo

df["Tipo_solicitud_1"] = df["Tipo_solicitud_1"].fillna("OTRA SOLICITUD")
```

```python
df
```

## 4. Preparar el Dataset quitando columnas o filas vacías y/o con valores que no corresponden

```python
# Cantidad de registros antes
print(f"Registros totales antes: {len(df)}")

# Eliminar aplicaciones que no corresponden
df = df[~df["Aplicacion"].isin(["liquidaciones de crédito de moneda extranjera", "SISCAR", "MONEXT"])]

# Cantidad de registros después
print(f"Registros totales después: {len(df)}")
```

```python
df = df.dropna(axis=1, how="all")
```

```python
df
```

## 5. Crear columnas nuevas según el Dataset actual

```python
festivos_colombia = holidays.Colombia(years=[2024, 2025, 2026])
def tipo_dia_calendario(fecha):
    fecha = fecha.date()
    if fecha.weekday() >= 5:      # 5 = sábado, 6 = domingo
        return "FESTIVO"
    if fecha in festivos_colombia: # día festivo en Colombia
        return "FESTIVO"
    return "HABIL"

df["Tipo_dia_apertura"] = df["Fecha_apertura"].apply(tipo_dia_calendario)
df["Tipo_dia_cierre"] = df["Fecha_cierre"].apply(tipo_dia_calendario)
```

```python
df["Nro_dia_apertura"] = df["Fecha_apertura"].dt.day
df["Nro_dia_cierre"] = df["Fecha_cierre"].dt.day
df["Mes_apertura"] = df["Fecha_apertura"].dt.month
df["Mes_cierre"] = df["Fecha_cierre"].dt.month
df["Hora_apertura"] = df["Fecha_apertura"].dt.hour
df["Hora_cierre"] = df["Fecha_cierre"].dt.hour

df["Año_caso"] = np.where(
    df["Fecha_apertura"].dt.year == df["Fecha_cierre"].dt.year,
    df["Fecha_apertura"].dt.year,
    df["Fecha_cierre"].dt.year
)
```

```python
df
```

```python
festivos = np.array(list(festivos_colombia.keys()), dtype='datetime64[D]')

df["ANS_dias_laborales"] = df.apply(
    lambda x: np.busday_count(
        x["Fecha_apertura"].date(),
        x["Fecha_cierre"].date(),
        holidays=festivos
    ),
    axis=1
)
```

```python
df
```

```python
def horario_laboral(fecha):
    if fecha.weekday() >= 5:
        return "SI"
    if fecha.hour >= 8 and fecha.hour < 18:
        return "NO"
    return "SI"

df["Apertura_fuera_horario"] = df["Fecha_apertura"].apply(horario_laboral)
df["Cierre_fuera_horario"] = df["Fecha_cierre"].apply(horario_laboral)
df["Cumple_ANS"] = df["ANS_dias_laborales"].apply(lambda x: "SI" if x <= 3 else "NO")
```
