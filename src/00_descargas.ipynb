## Extracción datos de carpeta compartida - Dataset Recursos Humanos

```python
from pathlib import Path
import shutil

ORIGEN = Path(r"A:\Universidad\AccesClean\RecursosHumanos")
DESTINO = Path(r"../data/raw/rrhh")

for archivo in ORIGEN.glob("*.xlsx"):
    print("[1] Archivo(s) a copiar en carpeta: "+archivo.name)
    
    if archivo.name.startswith("~$"):
        print("[2] Archivo temporal ignorado: "+archivo.name)
        continue
        
    shutil.copy2(archivo, DESTINO)
    print("[3] Archivo(s) copiados exitosamente en destino: "+archivo.name)
```
