## Datos extraidos de carpeta compartida - Dataset Recursos Humanos

```python
from pathlib import Path
import shutil

ORIGEN = Path(r"A:\Universidad\AccesClean\RecursosHumanos")
DESTINO = Path(r"../data/raw/rrhh")

if ORIGEN.exists():
    print("[1] ✓ Carpeta origen encontrada")
else:
    print("[1] ✗ Carpeta origen no encontrada")

for archivo in ORIGEN.glob("*"):
    print("[2] "+archivo.name)

for archivo in ORIGEN.glob("*"):
    shutil.copy2(archivo, DESTINO)
print("[3] Archivos copiados correctamente")

for archivo in DESTINO.glob("*"):
    print("[4] "+archivo.name)
```
